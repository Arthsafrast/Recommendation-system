# **Creating recommendation system with surprise library**

# **Content based recommendation system**

Stop_words = "english" means that all the stop words such as the, a ect. will be removed from descriptions. I also removed spaces from the descriptions and finally using tfidf transformed the text into vectors, which also give weights to every word by giving the rare term a high weight and giving the common term a low weight

```python
tfidf = TfidfVectorizer(stop_words='english')

df_cont['About Product'] = df_cont['About Product'].fillna('')
tfidf_matrix = tfidf.fit_transform(df_cont['About Product'])
tfidf_matrix.shape
```

Here I found the similarity matrix using linear_kernel function and also created a series that maps the index of the matrix to movie names.

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
indices = pd.Series(df_cont.index, index=df_cont['Name'])
indices
```

Here I made a recommender function that will recommend us movies using cosine_similarity. Our function will take an item name as input and then find the top 11 items using the cosine similarity matrix we found above.

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x:x[1], reverse=True)
    sim_scores = sim_scores[1:11]

    rec_indices = [i[0] for i in sim_scores]
    return df_cont['Name'].iloc[rec_indices]
```

Here we get the description of those recommended items based on the same principle.

```python
def get_recommendations_desc(title, cosine_sim=cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x:x[1], reverse=True)
    sim_scores = sim_scores[1:11]

    rec_indices = [i[0] for i in sim_scores]
    return df_cont['About Product'].iloc[rec_indices]
```
